# Arithmetic Series

## Modified from Arithmetic Series -  [Wolfram MathWorld](https://mathworld.wolfram.com/ArithmeticSeries.html) - [wikipedia](https://en.wikipedia.org/wiki/Arithmetic_progression)


An arithmetic sequence $\{a_k\}$ for $k=1, 2, ..., $ in which each term is computed from the previous one by adding (or subtracting) a constant $d$.

\begin{align*}
a_1 & \\
a_2 &\ \ =\ \ a_1 + d  \\
a_3 &\ \ =\ \ a_2 + d \ \ =\ \ a_1 + 2d   \\
a_4 &\ \ =\ \ a_3 + d \ \ =\ \ a_2 + 2d \ \ =\ \ a_1 + 3d  \\
\vdots\\
a_k &\ \ =\ \ a_{(k-1)}+d \ =\  a_{(k-2)}+2d \ =\  \ldots \ =\  \ a_1+(k-1)d
\end{align*}

Therefore, for $k>=1$,

$$ a_k = a_1 + (k-1)d$$

With $a_1$ and $d$ given.

<br />

The sequence of the partial sum of the first $n$ terms of a sequence is called a series, in the case of an arithemetic sequence the series is called an arithmetic series.

If  $a_1= 1$ and $d=1$ we have, the simplest arithmnetic sequence,  the natural numbers sequence $\{a_k = k \}$ for  $k=1, 2, ..., $  i.e. the seuqnece
$$1, 2, ...,$$
The  corresponding series is
$$
S_n = 1 + 2+ \ldots + (n-1)+ n = \sum_{k=1}^n k
$$
can be calculated as a function of $n$.

Assume that we allign one $S_{n-1}$ with a second  $S_{n-1}$ that have the terms adden reverse order and you add term to term, we have
\begin{array}{llclclclclc}
  S_{n-1} &=& 1 &+& 2 &+& \ldots &+& (n-2) &+& (n-1) \\
  S_{n-1} &=& (n-1)  &+& (n-2)  &+& \ldots   &+& 2   &+& 1 \\
   \hline \\
   2S_{n-1} &=& n  &+& n  &+& \ldots   &+& n   &+& n \\
\end{array}
so
$$  2S_{n-1} = (n-1)n $$
then
$$  S_{n-1} = \frac{1}{2} (n-1)n$$
therefore
$$  S_{n} = \frac{1}{2} n(n+1)$$


\begin{align*}
S_n &= \sum_{k=1}^n a_k	\\
    &= \sum_{k=1}^n[a_1+(k-1)d]		\\
    &= na_1 + d \sum_{k=1}^n(k-1)	\\
    &= na_1 + d \sum_{k=0}^{n-1} k	\\
    &= na_1 + d \sum_{k=1}^{n-1} k	\\
    &= na_1 +  \frac{1}{2} d(n-1)n\\
\end{align*}
so
$$ S_n =  \frac{1}{2} n[2 a_1 +d(n-1)].$$

Note, however, that
$$a_1+a_n=a_1+[a_1+d(n-1)]=2a_1+d(n-1),$$
so
$$S_n = \frac{1}{2}n(a_1+a_n),$$


or $n$ times the arithmetic mean of the first and last terms! This is the trick Gauss used as a schoolboy to solve the problem of summing the integers from $1$ to $100$ given as busy-work by his teacher. While his classmates toiled away doing the addition longhand, Gauss wrote a single number, the correct answer

$$\frac{1}{2}(100)(1+100)=50Â·101=5050$$
on his slate (Burton 1989, pp. 80-81; Hoffman 1998, p. 207). When the answers were examined, Gauss's proved to be the only correct one.

<br />

If  the sequence starts at index $0$, $\{a_k\}$ for $ k=0, 1, 2, ..., $ we have

\begin{align*}
a_0 & \\
a_1 &\ \ =\ \ a_0 + d  \\
a_2 &\ \ =\ \ a_1 + d \ \ =\ \ a_0 + 2d   \\
a_3 &\ \ =\ \ a_2 + d \ \ =\ \ a_1 + 2d \ \ =\ \ a_0 + 3d  \\
\vdots\\
a_k &\ \ =\ \ a_{(k-1)}+d \ =\  a_{(k-2)}+2d \ =\  \ldots \ = \  \ a_1+(k-1)d \ \ =\ \ a_0+kd
\end{align*}

<br />
for given   $a_0$ and $d$.

In this case the sum of the terms between $0$ and $n$ of  the arithmetic sequence  is then given by
\begin{align*}
S_n &= \sum_{k=0}^{n} a_k	\\
    &= \sum_{k=0}^{n}[a_0+kd]		\\
    &= (n+1) a_0 + d \sum_{k=0}^{n} k	\\
    &= (n+1) a_0 + d \sum_{k=1}^{} k	\\
    &=(n+1) a_0+  \frac{1}{2} dn(n+1) \\
\end{align*}
so
$$ S_n =  \frac{1}{2} (n+1)[2 a_0 + dn].$$
<br />
$$ S_n =  \frac{1}{2} (n+1)[a_0+ a_n].$$

# Geometric Series

## Modified from Geometric Series -  [Wolfram MathWorld](https://mathworld.wolfram.com/GeometricSeries.html) - [wikipedia](https://en.wikipedia.org/wiki/Geometric_series)


An geometric series is the sum of a sequence $\{a_k\}$ for $k=1, 2, ..., $ in which each term is computed from the previous  by multipliying constant $r$ and starts with 1.

\begin{align*}
a_0 &\ \ =\ \ r^0 \ \ =\ \ 1\\
a_1 &\ \ =\ \ a_0 *  r \ \ =\ \  r\\
a_2 &\ \ =\ \ a_1 * r\ \ =\ \ a_0 * r^2  \ \ =\ \ r^2 \\
a_3 &\ \ =\ \ a_2 * r \ \ =\ \ a_0 * r^3 \ \ =\ \  r^3  \\
a_4 &\ \ =\ \ a_3 * r\ \  =\ \ a_0 * r^4 \ \  =\ \ r^4  \\
\vdots\\
a_k &\ \ =\ \ a_{(k-1)}* r \ =\   a_0 * r^k \ \  =\ \  r^k
\end{align*}

The sum of the first $n$ terms  of  the geometric sequence (called an geometric serie)  is then given by
$$
\begin{array}{ll}
S_n &=  1 + r + r^2 + \cdots + r^{n-1} +r^n \\
 &=  (1 + r + r^2 + \cdots + r^{n-1} ) +r^n \\
 &= S_{n-1}  +r^n
\end{array}
$$

$$
\begin{array}{ll}
S_n &=  1 + r + r^2 + \cdots + r^{n-1} +r^n \\
 &=  1 + (r + r^2 + \cdots + r^{n-1}  +r^n)  \\
 &=   1 + r (1 +  r  + r^2 + \cdots + r^{n-1} )  \\
  &=   1 + r   S_{n-1}
\end{array}
$$

Then
$$
 S_{n-1}  +r^n =   1 + r   S_{n-1}
$$

Therefore
$$
 S_{n-1} (1 -r)   +r^n =   1 - r^n
$$

$$
 S_{n-1}   =   \frac{1 - r^n }{1 -r}
$$

$$
 S_{n}   =   \frac{1 - r^{n+1} }{1 -r}  = \frac{ r^{n+1} - 1}{r -1}
$$

### **Objetivo:**
1. Calcular la suma de la serie geomÃ©trica para \( n = 10^6 \)



```
import sys

# Definir valores iniciales
r = 2
n = 10**6  # NÃºmero de tÃ©rminos

# Calcular la suma con la fÃ³rmula
S_n_formula = (r**(n+1) - 1) // (r - 1)

# Increase the limit for integer string conversion
sys.set_int_max_str_digits(0) # Setting to 0 removes the limit.

print(S_n_formula)
# Mostrar solo la longitud del nÃºmero para evitar problemas de memoria
print(f"La suma tiene aproximadamente {len(str(S_n_formula))} dÃ­gitos.")
```

## Plot and  print the first 20 terms and the partial sums.

```
import numpy as np
import matplotlib.pyplot as plt

# Definir valores iniciales
r = 2
n = 20  # NÃºmero de tÃ©rminos

# Generar la secuencia geomÃ©trica y las sumas parciales
sequence_a_n = [r**k for k in range(n)]  # TÃ©rminos de la serie
partial_sums_S_n = np.cumsum(sequence_a_n)  # Sumas parciales

# Imprimir los tÃ©rminos y sus sumas parciales
print("TÃ©rmino (a_n)   Suma Parcial (S_n)")
for i in range(n):
    print(f"{sequence_a_n[i]:<15} {partial_sums_S_n[i]}")

# Graficar los resultados
plt.figure(figsize=(10, 5))
plt.plot(range(n), sequence_a_n, 'ro-', label='TÃ©rminos (a_n)')
plt.plot(range(n), partial_sums_S_n, 'bo-', label='Sumas parciales (S_n)')
plt.xlabel("n")
plt.ylabel("Valor")
plt.title("Serie GeomÃ©trica y Sumas Parciales")
plt.legend()
plt.grid(True)
plt.show()
```

## Sum the  geometric with  $r=\frac{1}{2}$ add the series up to 10**3 and verify with the formula.



#Suma de una Serie GeomÃ©trica con $r = \frac{1}{2}$

La serie geomÃ©trica dada por:

$$
S_n = \sum_{k=0}^{n} \left(\frac{1}{2}\right)^k
$$

puede calcularse usando la fÃ³rmula de la suma de una serie geomÃ©trica:

$$
S_n = \frac{1 - r^{n+1}}{1 - r}
$$

Para $ r = \frac{1}{2} $, esto se convierte en:

$$
S_n = \frac{1 - \left(\frac{1}{2}\right)^{n+1}}{1 - \frac{1}{2}} = 2 \left(1 - \left(\frac{1}{2}\right)^{n+1} \right)
$$

Cuando $( n \to \infty )$ la suma converge a:

$$
S = \frac{1}{1 - \frac{1}{2}} = 2
$$

Verificaremos esto sumando la serie hasta \( n = 10^3 \) y comparando con la fÃ³rmula.


```
import numpy as np

# ParÃ¡metros de la serie
r = 1/2
n = 10**3  # NÃºmero de tÃ©rminos

# Calcular la suma iterativa
S_n_iterative = sum(r**k for k in range(n))

# Calcular la suma con la fÃ³rmula
S_n_formula = (1 - r**(n+1)) / (1 - r)

# Imprimir los resultados
print(f"Suma iterativa: {S_n_iterative}")
print(f"Suma con fÃ³rmula: {S_n_formula}")
print(f"Coinciden: {np.isclose(S_n_iterative, S_n_formula)}")
```

## Your own example of power sum with $p =$ last digit in the list $+ 1$, i.e., the sum $1^p+2^p+\cdots + n^p$  series up to  $n=10^6$  and verify with the formula.


# La suma de potencias de los primeros $n$ enteros positivos se define como:

$$
S_p(n) = 1^p + 2^p + 3^p + \dots + n^p = \sum_{k=1}^{n} k^p
$$

donde $p$ es un exponente entero. En este caso, tomaremos $p$ como el Ãºltimo dÃ­gito de nuestra lista mÃ¡s uno.

#FÃ³rmula de la Suma de Potencias
La suma de potencias tiene una representaciÃ³n en tÃ©rminos de los **nÃºmeros de Bernoulli** $B_k$:

$$
S_p(n) = \frac{1}{p+1} \sum_{k=0}^{p} \binom{p+1}{k} B_k n^{p+1-k}
$$

Esta expresiÃ³n nos permite calcular eficientemente la suma sin necesidad de iterar por todos los valores hasta $n$.



*   Calcularemos la suma hasta $n=10^6$ tanto iterativamente como usando la fÃ³rmula y verificaremos que los resultados coinciden.




```
import numpy as np
import sympy as sp

# Definir valores
n = 10**6  # LÃ­mite superior de la serie
p = 2  # Ejemplo: p = Ãºltimo dÃ­gito en la lista + 1 (ajÃºstalo segÃºn tu necesidad)

# CÃ¡lculo iterativo
S_p_iterative = sum(k**p for k in range(1, n+1))

# CÃ¡lculo con la fÃ³rmula usando nÃºmeros de Bernoulli
B = [sp.bernoulli(k) for k in range(p+1)]  # NÃºmeros de Bernoulli hasta p
S_p_formula = sum(sp.binomial(p+1, k) * B[k] * n**(p+1-k) for k in range(p+1)) / (p+1)

# VerificaciÃ³n
print(f"Suma iterativa: {S_p_iterative}")
print(f"Suma con fÃ³rmula: {int(S_p_formula)}")  # Convertimos a entero para evitar decimales
print(f"Coinciden: {S_p_iterative == int(S_p_formula)}")
```
